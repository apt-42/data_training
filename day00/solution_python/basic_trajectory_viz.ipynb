```python
# import requirements
import pandas as pd # analysis
import plotly.express as px # graphing
```

```python
# path to dataset
df_path = 'data.csv'
```

```python
# create a Pandas DataFrame object from a CSV file
df = pd.read_csv(df_path)
```

### Quick inspection

```python
# use pandas.DataFrame.columns to display the colums name
df.columns
```

```python
# display the first rows
df.head(5)
# and the last ones
# df.tail()
```

```python
# select useful columns
df = pd.read_csv(df_path, usecols=[0,1,2], float_precision='high')
df.head()
```

### Vizualisation

```python
fig = px.scatter(df,
                 df['longitude'],
                 df['latitude'],
                 hover_data={
                         'longitude': ':.15f',
                         'latitude': ':.15f',
                     }
                 )

fig.show()
```

longitude = 2.349234411565736
latitude = 48.859425557543304

```python
# yes!
df[(df['longitude'] == 2.349234411565736)]
```

```python
# nothing!
df[(df['latitude'] == 48.859425557543304)]
```

```python
# then, boolean mask to handle floating point rounding
df[(df['longitude'] > 2.349234411565735) & (df['longitude'] < 2.349234411565737)]
```
